```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pywt
import pywt.data
```

```python
img = mpimg.imread ('formula1.jpg')
if img.ndim == 3:
    # Convertir a escala de grises
    img = np.dot(img[..., :3], [0.2989, 0.5870, 0.1140])
plt.imshow(img, cmap='gray')
print(img.shape)
```

```python
coeffs = pywt.wavedec2(img, 'db1', mode='periodization', level=2)
cA2, (cH2, cV2, cD2), (cH1, cV1, cD1) = coeffs  
plt.figure(figsize=(12, 12))
plt.subplot(221), plt.imshow(cA2, cmap='gray'), plt.title('Approximation Coefficients cA2')
plt.subplot(222), plt.imshow(cH2,cmap= 'gray')
plt.show()
```
